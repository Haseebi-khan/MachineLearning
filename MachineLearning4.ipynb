### **Multi Linear Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv(r"D:\Codes\MachineLearning\Datasets\advertising4D.csv")
df
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
tv = np.array(df["TV"])
newspaper = np.array(df["newspaper"])
radio = np.array(df["radio"])
y = np.array(df["sales"])


x = tv
numerator = np.sum(np.multiply(x - np.mean(x), y - np.mean(y)))
denominator = np.multiply(np.sqrt(np.sum((x - np.mean(x))**2)), np.sqrt(np.sum((y - np.mean(y))**2)))
corr_coef_tv = numerator / denominator

x = radio
numerator = np.sum(np.multiply(x - np.mean(x), y - np.mean(y)))
denominator = np.multiply(np.sqrt(np.sum((x - np.mean(x))**2)), np.sqrt(np.sum((y - np.mean(y))**2)))
corr_coef_radio = numerator / denominator

x = newspaper
numerator = np.sum(np.multiply(x - np.mean(x), y - np.mean(y)))
denominator = np.multiply(np.sqrt(np.sum((x - np.mean(x))**2)), np.sqrt(np.sum((y - np.mean(y))**2)))
corr_coef_newspaper = numerator / denominator

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 6))
fig.suptitle('Pearson Correlation Coefficient', fontsize=16)

ax1.set_title(f'Pearson Correlation Coefficient: {corr_coef_tv:.3f}')
sns.scatterplot(x='TV', y='sales', data=df, ax=ax1)

ax2.set_title(f'Pearson Correlation Coefficient: {corr_coef_radio:.3f}')
sns.scatterplot(x='radio', y='sales', data=df, ax=ax2)

ax3.set_title(f'Pearson Correlation Coefficient: {corr_coef_newspaper:.3f}')
sns.scatterplot(x='newspaper', y='sales', data=df, ax=ax3)

plt.tight_layout()
plt.show()
```

### **Creating X Matix and Y Vector.**

```python
y = np.array(df["sales"])
y
```

### **Creating X Matrix:**

```python
# axis=0 (or axis='index') refers to the rows. Operations performed with axis=0 will typically act across each column.   
# axis=1 (or axis='columns') refers to the columns. Operations performed with axis=1 will typically act across each row.   

features = np.array(df.drop("sales", axis=1))
features
```

```python
X = np.insert(features, 0, 1, axis=1)
X
```

```python
rows = df["radio"].shape[0]
ones = np.ones([rows, 1])
X = np.hstack((ones,features))
X
```

### **Calculating Regression Coefficients.**

```python
# X Transpose
# X.T@X    
np.dot(X.T, X)
```

```python
# Finding Inverse
np.linalg.inv(np.dot(X.T, X))
```

```python
# Transpose with Y
np.dot(X.T, y)
# X.T@y    
```

```python
# formula value such b0, b1, bi will be Computed below

# values = np.linalg.inv(np.dot(X.T, X)) @ np.dot(X.T, y)

beta = np.linalg.inv(X.T @ X) @ (X.T @ y)

print("Beta0: {:.3f},      Beta1: {:.3f},         Beta2: {:.3f},          Beta3: {:.3f}".format(beta[0],beta[1],beta[2],beta[3]))


```

```python
_tv        = 180
_radio     = 10
_newspaper = 58
predict_Value = beta[0] + (beta[1] * _tv) +  (beta[2] * _radio)  + (beta[3] * _newspaper)

print("predict_Value for sales is: {:.3f}".format(predict_Value))
```

```python
ybar = np.mean(y)

# Assuming you have calculated your betas (coefficients)
# Replace these with your actual calculated beta values
yhat = beta[0] + beta[1] * df['TV'] + beta[2] * df['radio'] + beta[3] * df['newspaper']

mae = np.mean(np.abs(y - yhat))
mse = np.mean((y - yhat)**2)
rmse = np.sqrt(np.mean((y - yhat)**2))
sse = np.sum((y - yhat)**2)
sst = np.sum((y - ybar)**2)
r2 = 1 - sse/sst

# Assuming 'features' is your DataFrame without the target variable ('sales' in the original context)
features = df.drop('sales', axis=1)
n = len(y)
p = features.shape[1] # Number of features

r2a = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"R2: {r2:.3f}")
print(f"Adjusted R2: {r2a:.3f}")
```
